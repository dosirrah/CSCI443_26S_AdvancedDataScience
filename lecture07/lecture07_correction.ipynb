### Why not use sum of deviations as a measure of dispersion?

$$
\frac{1}{n} \sum_{i=1}^n (X_i - \bar{X})
$$

*Answer:*

The answer, "it is zero."  That is absolutely correct.

$$
\begin{aligned}
\frac{1}{n} \sum_{i=1}^n (X_i - \bar{X}) &=
\left(\frac{1}{n} \sum_{i=1}^n X_i \right) - \left(\frac{1}{n} \sum_{i=1}^n \bar{X} \right) \\
 &= \bar{X} - \frac{1}{n} \cdot n\bar{X} \\
 &= \bar{X} - \bar{X} \\
 &= 0
\end{aligned}
$$


With Mean Absolute Deviation, there are no negative terms to cancel positive terms.

$$
\frac{1}{n} \sum_{i=1}^n \left| X_i - \bar{X} \right|
$$

Mean Absolute Deviation is sometimes preferred over standard deviation because 
it less sensitive to large errors.  However it is less robust than 
Median Absolute Deviation 

The fact that Mean Absolute Deviation shares the first letter of each word 
with Mean Absolute Deviation is confusing.  They are separate concepts.
MAD usually refers to Median Absolute Deviation (MAD).  

*Practical Statistics for Data Scientists* uses the name 
"Median Absolute Deviation from the Median" though it is more commonly
called Median Absolute Deviation (MAD).

$$
\text{MAD} = \text{median}(|x_i - \text{median}(x)|)
$$
