Generate some p-values

```python
import random
import pandas as pd
random.seed(17)

def generate_p_values():
    p_values = []
    count_below_5 = 0

    # Keep generating values until we have 7 below 0.05
    while count_below_5 < 7:
        p = random.uniform(0, 1)
        p_values.append(p)
        if p < 0.05:
            count_below_5 += 1

    # Sort the list in increasing order
    p_values.sort()
    
    return p_values

# Generate and display the sorted p-values
sorted_p_values = generate_p_values()

# Create a DataFrame for pretty display
df = pd.DataFrame({'p-value': sorted_p_values})

# Display the table in a Jupyter-friendly format
#import ace_tools as tools
#tools.display_dataframe_to_user(name="Sorted P-Values", dataframe=df_sorted_p_values)

#df.head(100)

df = df.iloc[0:100]
print(df.count())

from IPython.display import display
display(df.iloc[0:10])
display(df)
```

```python
# Number of hypotheses (total p-values)
m = len(df)

# Set FDR threshold
FDR = 0.05

# Compute alpha_i values using Benjamini-Hochberg procedure
df = df.sort_values(by='p-value')  # Ensure sorting
df['alpha (FDR 5%)'] = [(i+1)/m * FDR for i in range(m)]

# Determine significance: p-value < alpha_i (FDR threshold per rank)
#df_sorted_p_values['Significant (FDR 5%)'] = df_sorted_p_values['p-value'] < df_sorted_p_values['alpha (FDR 5%)']

display(df.iloc[0:10])
display(df)
```
